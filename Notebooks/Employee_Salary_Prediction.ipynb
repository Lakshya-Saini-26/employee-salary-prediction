```python
X=df.drop('Salary',axis=1)
y=df['Salary']
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
X['Gender'] = le.fit_transform(X['Gender'])
X['Education Level'] = le.fit_transform(X['Education Level'])
```

```python
# from sklearn.preprocessing import OneHotEncoder
# enc=OneHotEncoder(handle_unknown='ignore',sparse_output=False)
# enc_train=enc.fit_transform(X_train[['Job Title']])
# enc_test=enc.transform(X_test[['Job Title']])
```

```python
X=pd.get_dummies(X,drop_first=True)
```

```python
X=X.astype(int)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.20, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
model_LR=LinearRegression()
model_LR.fit(X_train,y_train)
```

```python
y_pred_LR=model_LR.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
r2 = r2_score(y_test, y_pred_LR)
mae = mean_absolute_error(y_test, y_pred_LR)
mse = mean_squared_error(y_test, y_pred_LR)
rmse = np.sqrt(mse)

print(f"R-squared: {r2:.2f}")
print(f"MAE: {mae:.2f}")
print(f"RMSE: {rmse:.2f}")
```

```python
from sklearn.neighbors import KNeighborsRegressor
model_KNN=KNeighborsRegressor(n_neighbors=5)
model_KNN.fit(X_train,y_train)
```

```python
y_pred_KNN=model_KNN.predict(X_test)
```

```python
r2 = r2_score(y_test, y_pred_KNN)
mae = mean_absolute_error(y_test, y_pred_KNN)
mse = mean_squared_error(y_test, y_pred_KNN)
rmse = np.sqrt(mse)

print(f"R-squared: {r2:.2f}")
print(f"MAE: {mae:.2f}")
print(f"RMSE: {rmse:.2f}")
```

```python
from sklearn.naive_bayes import GaussianNB
model_NB = GaussianNB()
model_NB.fit(X_train,y_train)
```

```python
y_pred_NB=model_NB.predict(X_test)
```

```python
r2 = r2_score(y_test, y_pred_NB)
mae = mean_absolute_error(y_test, y_pred_NB)
mse = mean_squared_error(y_test, y_pred_NB)
rmse = np.sqrt(mse)

print(f"R-squared: {r2:.2f}")
print(f"MAE: {mae:.2f}")
print(f"RMSE: {rmse:.2f}")
```

```python
from sklearn.svm import SVR
model_SVR=SVR()
model_SVR.fit(X_train,y_train)
```

```python
y_pred_SVR=model_SVR.predict(X_test)
```

```python
r2 = r2_score(y_test, y_pred_SVR)
mae = mean_absolute_error(y_test, y_pred_SVR)
mse = mean_squared_error(y_test, y_pred_SVR)
rmse = np.sqrt(mse)

print(f"R-squared: {r2:.2f}")
print(f"MAE: {mae:.2f}")
print(f"RMSE: {rmse:.2f}")
```

```python
from sklearn.tree import DecisionTreeRegressor
model_DT=DecisionTreeRegressor()
model_DT.fit(X_train,y_train)
```

```python
y_pred_DT=model_DT.predict(X_test)
```

```python
r2 = r2_score(y_test, y_pred_DT)
mae = mean_absolute_error(y_test, y_pred_DT)
mse = mean_squared_error(y_test, y_pred_DT)
rmse = np.sqrt(mse)

print(f"R-squared: {r2:.2f}")
print(f"MAE: {mae:.2f}")
print(f"RMSE: {rmse:.2f}")
```

```python
from sklearn.ensemble import RandomForestRegressor
model_RF=RandomForestRegressor(random_state=42)
model_RF.fit(X_train,y_train)
```

```python
y_pred_RF=model_RF.predict(X_test)
```

```python
r2 = r2_score(y_test, y_pred_RF)
mae = mean_absolute_error(y_test, y_pred_RF)
mse = mean_squared_error(y_test, y_pred_RF)
rmse = np.sqrt(mse)

print(f"R-squared: {r2:.2f}")
print(f"MAE: {mae:.2f}")
print(f"RMSE: {rmse:.2f}")
```

```python
print("Training R²:", model_RF.score(X_train, y_train))
print("Testing R² :", model_RF.score(X_test, y_test))
```

```python
from sklearn.model_selection import GridSearchCV
model_RFCV=RandomForestRegressor()
parameters = {
    'n_estimators':[100,500,1000],
    'bootstrap':(True,False),
    'max_depth':[10,20,None],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
        }
RF_reg=GridSearchCV(model_RFCV,parameters,cv=5)
RF_reg.fit(X_train,y_train)
```

```python
y_pred_RF_reg=RF_reg.predict(X_test)
```

```python
r2 = r2_score(y_test, y_pred_RF_reg)
mae = mean_absolute_error(y_test, y_pred_RF_reg)
mse = mean_squared_error(y_test, y_pred_RF_reg)
rmse = np.sqrt(mse)

print(f"R-squared: {r2:.2f}")
print(f"MAE: {mae:.2f}")
print(f"RMSE: {rmse:.2f}")
```
